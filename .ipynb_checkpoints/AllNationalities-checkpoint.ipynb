# Data analysis births among internatiol football players

The Fifa data set is a set on kaggle where all the data of the game is stored. The data set can be found in following link:

https://www.kaggle.com/stefanoleone992/fifa-21-complete-player-dataset/version/1

### import data

```python
# import liberaries
import pandas as pd

players = pd.read_excel("Career Mode player datasets - FIFA 15-21.xlsx", index_col = None)
```

```python
players.head(10)
```

```python
print(players.shape)
num_players = len(players)
num_players
```

```python
from pandas import Timestamp
players['birth_month'] = players['dob'].apply(lambda x: x.month)
players['birth_month_written'] = players['dob'].apply(lambda x: x.strftime("%B"))

players['birth_year'] = players['dob'].apply(lambda x: x.year)
players.head(10)
```

```python
# how many belgians
begian_players = players[players['nationality'] == "Belgium"]
begian_players
```

```python
len(begian_players)
```

```python
birth_month_players = players.groupby(['birth_month_written']).count()
birth_month_players

months = ['January', 'February', 'March', 'April', 'May', 'June','July','September','October','November','December']
```

```python
birth_month_players = birth_month_players.reindex(index = months)
birth_month_players
```

```python
birth_month_players = birth_month_players.reset_index()
birth_month_players['pourcentage_births'] = birth_month_players['sofifa_id']/num_players
birth_month_players
```

```python
import matplotlib.pyplot as plt

plt.plot(birth_month_players['birth_month_written'], birth_month_players['sofifa_id'],marker = 'o', linestyle = '--')

plt.ylabel("montly births")
plt.xlabel("")
plt.xticks(rotation=90)
plt.title("number of montlhy births among international players")
plt.ylim(0,1900)

plt.show()
```

```python
plt.plot(birth_month_players['birth_month_written'], birth_month_players['pourcentage_births'],marker = 'o', linestyle = '--')

plt.ylabel("pourcentage of montly births")
plt.xlabel("")
plt.xticks(rotation=90)
plt.title("pourcentage of montlhy births among international players")
plt.ylim(0,0.15)

plt.show()
```

## Belgian footballers

### Clean data

```python
import pandas as pd

Data_Belgian_Players = pd.read_json("Data_Belgian_Players.json")
num_Belgian_players = len(Data_Belgian_Players)
Data_Belgian_Players.head(10)
```

```python
Data_Belgian_Players["birth_year"]= Data_Belgian_Players["Date of Birth"].apply(lambda x: x.lstrip().split(" ")[2])
Data_Belgian_Players["birth_month"]= Data_Belgian_Players["Date of Birth"].apply(lambda x: x.lstrip().split(" ")[1])
Data_Belgian_Players["birth_day"]= Data_Belgian_Players["Date of Birth"].apply(lambda x: x.lstrip().split(" ")[0])
Data_Belgian_Players
```

```python
gb = Data_Belgian_Players.groupby(['birth_month']).count()
gb
```

```python
import matplotlib.pyplot as plt

months = ['January', 'February', 'March', 'April', 'May', 'June','July','September','October','November','December']
```

```python
x = gb['firstname']
```

```python
# put them in the right order
x = x.reindex(index = months)


x.plot(marker = 'o', linestyle = '--')

plt.ylabel("montly births")
plt.xlabel("")
plt.title("number of montlhy births among elite belgian players")
plt.ylim(0,140)

plt.show()
```

```python
# percentage
gb = gb.reindex(index = months)
gb = gb.reset_index()
gb['pourcentage_births'] = gb['firstname']/num_Belgian_players
gb
```

```python
plt.plot(gb['birth_month'], gb['pourcentage_births'],marker = 'o', linestyle = '--')

plt.ylabel("pourcentage of montly births")
plt.xlabel("")
plt.xticks(rotation=90)
plt.title("pourcentage of montlhy births among Belgian players")
plt.ylim(0,0.15)

plt.show()
```

```python
# combine both graphs

plt.plot(birth_month_players['birth_month_written'], birth_month_players['pourcentage_births'],marker = 'o', linestyle = '--')
plt.plot(gb['birth_month'], gb['pourcentage_births'],marker = 'o', linestyle = '--', color='green')


plt.ylabel("pourcentage of montly births")
plt.xlabel("")
plt.xticks(rotation=90)
plt.title("pourcentage of montlhy births among international players")
plt.ylim(0,0.15)
plt.legend(["international footballers", "belgian footballers"], loc ="lower right")

plt.show()
```
